```python
import numpy as np
import pandas as pd
```

```python
# Environment Impact: E
# Usage and Relevance: U
# Cost and Business Impact: C

# Define the column names
columns = ['Energy Efficiency Rating (E)',
           'Resource Consumption (E)',
           'Data Center Location (E)',
           'Maintenance Cost (C)',
           'Obsolete Functionality (U)',
           'Redundancy (U)',
           'Low Usage (U)',
           'User Impact (C)',
           'Business Value (C)',
           'Compliance and Regulatory (U)'
           ]

# Create an empty dataframe with the specified columns
df = pd.DataFrame(columns=columns)

# generate random data for the dataframe of 0 or 1 for each column
for i in range(300):
    # Generate random data for the dataframe of 0 or 1 for each column
    np.random.seed(42)
    random_values = np.random.randint(0, 2, size=(300, len(columns)))

    # Create a new dataframe with the random values
    df = pd.DataFrame(random_values, columns=columns)

# Print the dataframe
df
```

```python
# based on df, we can calculate the total score for each row
# Total Score = sum of all columns
df['Total Score'] = df.sum(axis=1)
df
```

```python
#rank the rows based on the total score
df = df.sort_values(by='Total Score', ascending=False)
df
```

```python
# rank the rows based on the total score and sequence of columns (column 0 is the most important and column 9 is the least important)
df['Total Score'] = df['Energy Efficiency Rating (E)'] + df['Resource Consumption (E)'] + df['Data Center Location (E)'] + df['Maintenance Cost (C)'] + df['Obsolete Functionality (U)'] + df['Redundancy (U)'] + df['Low Usage (U)'] + df['User Impact (C)'] + df['Business Value (C)'] + df['Compliance and Regulatory (U)']
df = df.sort_values(by='Total Score', ascending=False)
df
```

```python
# create new column as TARGET and assign 1 if the total score is greater than 6, otherwise 0 and I want only the top 100 rows based on the total score to be 1 if the TARGET count is more than 100
# df['TARGET'] = np.where(df['Total Score'] > 6, 1, 0)
# df['TARGET'] = np.where(df['TARGET'].cumsum() > 100, 0, df['TARGET'])
df['Total Score'] = df['Energy Efficiency Rating (E)'] + df['Resource Consumption (E)'] + df['Data Center Location (E)'] + df['Maintenance Cost (C)'] + df['Obsolete Functionality (U)'] + df['Redundancy (U)'] + df['Low Usage (U)'] + df['User Impact (C)'] + df['Business Value (C)'] + df['Compliance and Regulatory (U)']
df = df.sort_values(by='Total Score', ascending=False).reset_index(drop=True)

# create new column as TARGET and assign 1 if the total score is greater than 6, otherwise 0 and I want only the top 100 rows based on the total score to be 1 if the TARGET count is more than 100
df['TARGET'] = np.where((df['Total Score'] > 5) & (df.index <= 100), 1, 0)
df
```

```python
df['TARGET'].value_counts()
```

```python
df.to_csv('data.csv', index=False)
```
